```python
# importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# data collection/loading the data
df = pd.read_csv("diabetes.csv")
df
```

```python
df.info()
```

```python
# to sum the number of null values in each column (isna - is null)
df.isna().sum()
```

```python
df.shape 
# this means there are 768 rows and 9 columns 
```

```python
# since there are some columns that have null values at `some points, so we can fill it like this 
# in this one we are basically filling the null values at any point in Insulin with the mean of all the values in Insulin 
df['Insulin'] = df['Insulin'].fillna(df['Insulin'].mean())
```

```python
# to make it easier we can fill for all the data points that have null in the whole dataset with the for loop 
to_fill = ['BloodPressure','SkinThickness','BMI']
for x in to_fill:
    df[x] = df[x].fillna(df[x].median())
```

```python
# now lets check the whole dataset to check for null values 
df.isna().sum()
```

```python
# for some other values that the null values are little we can drop the rows that have null values 
# in this case we are dropping the remaining one, which is Glucose
# the "inplace = True" basically makes the drop a permanent change !
df.dropna(inplace = True)
```

```python
# now lets check again 
df.isna().sum()
```
