```python
import tkinter as tk
from tkinter import messagebox as msgbox
import csv

# Global list to hold departments
departments_list = []

# Load departments from CSV
def load_departments():
    global departments_list
    departments_list = []
    with open("GIG-logistics.csv", "r") as file:
        reader = csv.reader(file)
        next(reader)  # skip header
        for row in reader:
            dept = row[2].strip()  # adjust column if needed
            if dept not in departments_list:
                departments_list.append(dept)

# Welcome window if employee exists
def welcomeMessage(name, department, members):
    window = tk.Toplevel(root)
    window.title("Employee Verified")
    window.geometry("400x300")

    tk.Label(window, text=f"Welcome {name}", font=("Arial", 14)).pack(pady=5)
    tk.Label(window, text=f"Department: {department}", font=("Arial", 12)).pack(pady=5)
    tk.Label(window, text="Other members in this department:", font=("Arial", 12)).pack(pady=5)

    # Listbox for colleagues
    listbox = tk.Listbox(window, width=50, height=10)
    listbox.pack(pady=5)

    if members:
        for member in members:
            listbox.insert(tk.END, member)
    else:
        listbox.insert(tk.END, "No other members in this department.")

# Verify employee
def verify_employee():
    name = name_entry.get().strip().lower()  # case-insensitive
    department = dept_var.get().strip()

    if not name or department == 'Select Department':
        msgbox.showwarning("Input Error", "Please enter both Name and select Department.")
        return

    found = False
    members = []

    with open("GIG-logistics.csv", "r") as file:
        reader = csv.reader(file)
        next(reader)
        for row in reader:
            row = [col.strip() for col in row]
            emp_name = row[1].lower()  # case-insensitive
            emp_dept = row[2]

            if emp_dept.lower() == department.lower() and emp_name != name:
                members.append(row[1])  # original case for display

            if emp_name == name and emp_dept.lower() == department.lower():
                found = True

    if found:
        welcomeMessage(name_entry.get().strip(), department, members)
    else:
        msgbox.showerror("Employee Check", "Employee does not exist.")

# Clear input fields
def clear_fields():
    name_entry.delete(0, tk.END)
    dept_var.set('Select Department')

# Open verification window
def open_verification():
    load_departments()  # load departments from CSV

    window = tk.Toplevel(root)
    window.title("Employee Verification")
    window.geometry("400x250")

    global name_entry, dept_var

    tk.Label(window, text="Enter Name:").pack(pady=5)
    name_entry = tk.Entry(window)
    name_entry.pack(pady=5)

    tk.Label(window, text="Select Department:").pack(pady=5)
    dept_var = tk.StringVar(window)
    dept_var.set('Select Department')
    dept_menu = tk.OptionMenu(window, dept_var, *departments_list)
    dept_menu.pack(pady=5)

    tk.Button(window, text="Verify", command=verify_employee).pack(pady=5)
    tk.Button(window, text="Clear", command=clear_fields, fg="white", bg="red").pack(pady=5)

# Main window
root = tk.Tk()
root.title("GIG Logistics Employee System")
root.geometry("300x150")

tk.Label(root, text="Welcome to GIG Logistics Employee System", font=("Arial", 12)).pack(pady=10)
tk.Button(root, text="Start Verification", command=open_verification, fg="white", bg="black").pack(pady=10)

root.mainloop()
```
