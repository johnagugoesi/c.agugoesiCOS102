```python
import tkinter as tk
from tkinter import messagebox as msgbox
import csv


# Welcome window if employee exists
def welcomeMessage(name, department, members):

    window = tk.Toplevel(root)
    window.title("Employee Verified")
    window.geometry("400x200")

    label1 = tk.Label(window, text=f"Welcome {name}")
    label1.pack()

    label2 = tk.Label(window, text=f"Department: {department}")
    label2.pack()

    label3 = tk.Label(window, text="Other members in this department:")
    label3.pack()

    label4 = tk.Label(window, text=", ".join(members))
    label4.pack()


# Function to verify employee
def verify_employee():

    name = name_entry.get()
    department = dept_entry.get()

    found = False
    members = []

    with open("GIG-logistics.csv", "r") as file:
        reader = csv.reader(file)

        for row in reader:

            emp_name = row[0]
            emp_dept = row[1]

            if emp_dept.lower() == department.lower():
                members.append(emp_name)

            if emp_name.lower() == name.lower() and emp_dept.lower() == department.lower():
                found = True

    if found:
        welcomeMessage(name, department, members)

    else:
        msgbox.showerror("Employee Check", "Employee does not exist.")


# Function to open verification window
def open_verification():

    msgbox.showinfo("Info", "Welcome to GIG Logistics Verification")

    window = tk.Toplevel(root)
    window.title("Employee Verification")
    window.geometry("400x200")

    global name_entry
    global dept_entry

    name_label = tk.Label(window, text="Enter Name")
    name_label.pack()

    name_entry = tk.Entry(window)
    name_entry.pack()

    dept_label = tk.Label(window, text="Enter Department")
    dept_label.pack()

    dept_entry = tk.Entry(window)
    dept_entry.pack()

    verify_button = tk.Button(window, text="Verify", command=verify_employee)
    verify_button.pack()


# Main window
root = tk.Tk()
root.title("Home Page")
root.geometry("300x150")

label = tk.Label(root, text="GIG Logistics Employee System\n")
label.pack()

button = tk.Button(root, text="Start Verification", command=open_verification)
button.pack()

button.config(fg="white", bg="black")

root.mainloop()
```
