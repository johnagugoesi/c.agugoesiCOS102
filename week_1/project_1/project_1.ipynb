# Simple Interest Flowchart
![Simple Interest Flowchart](simpleinterest.png)

```python
# Simple Interest
# Input
P = float(input("Enter Principal (P): "))
R = float(input("Enter Rate (%): "))
T = float(input("Enter Time (years): "))

#Calculate
A = P * (1 + (R / 100) * T)

#Result of the Simple Interest
print("Total Amount =", A)
```

# Compound Interest Flowchart
![Compound Interest Flowchart](compoundinterest.png)

```python
# Compound Interest 
# Input
P = float(input("Enter Principal (P): "))
R = float(input("Enter Annual Rate (%): ")) / 100  # convert percentage to decimal
n = int(input("Enter number of times compounded per year: "))
t = float(input("Enter Time in years: "))

# Calculation
A = P * (1 + R/n) ** (n * t)

# Result
print("Total Amount =", A)
```

# Annuity Plan Flowchart
![Annuity Plan Flowchart](annuityplan.png)

```python
# Annuity Plan

# Inputs
PMT = float(input("Enter Payment per period (PMT): "))
R = float(input("Enter Annual Rate (%): ")) / 100  # convert % to decimal
n = int(input("Enter number of payments per year: "))
t = float(input("Enter Time in years: "))

# Calculate
A = PMT * (((1 + R/n) ** (n * t) - 1) / (R/n))

# Output
print("Future Value of Annuity =", A)
```
