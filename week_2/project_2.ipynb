```python
print("ROOT FINDER PROGRAM")

print("1. Quadratic Equation")
print("2. Cubic Equation")
print("3. Quartic Equation")

choice = int(input("Choose operation (1-3): "))


# to calculate quadratic
if choice == 1:

    print("Quadratic Equation: Ax^2 + Bx + C = 0")

    A = float(input("Enter A: "))
    B = float(input("Enter B: "))
    C = float(input("Enter C: "))

    D = (B * B) - (4 * A * C)

    root1 = (-B + (D ** 0.5)) / (2 * A)
    root2 = (-B - (D ** 0.5)) / (2 * A)

    print("Root 1 =", root1)
    print("Root 2 =", root2)



# to calculate cubic
if choice == 2:

    print("Cubic Equation: Ax^3 + Bx^2 + Cx + D = 0")

    A = int(input("Enter A: "))
    B = int(input("Enter B: "))
    C = int(input("Enter C: "))
    D = int(input("Enter D: "))

    for x in range(-100, 101):

        y = A*(x**3) + B*(x**2) + C*x + D

        if y == 0:
            print("Root =", x)



# to calculate quartic
if choice == 3:

    print("Quartic Equation: Ax^4 + Bx^3 + Cx^2 + Dx + E = 0")

    A = int(input("Enter A: "))
    B = int(input("Enter B: "))
    C = int(input("Enter C: "))
    D = int(input("Enter D: "))
    E = int(input("Enter E: "))

    for x in range(-100, 101):

        y = A*(x**4) + B*(x**3) + C*(x**2) + D*x + E

        if y == 0:
            print("Root =", x)
```
